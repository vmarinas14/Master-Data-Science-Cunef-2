El dataset `BlackFriday.csv` contiene 550.000 registros sobre las ventas realizadas en un comercio. Cada registro contiene información del cliente (sexo, si está casado o no, edad, etc.) e información de la compra que ha realizado en el comercio (importe de la compra, tipo de producto, etc.).

* `User_ID`                : Código de usuario
* `Product_ID`             : Código de producto
* `Sexo`                   : Hombre(M), Mujer(F)
* `Edad`                   : Franja de edad del usuario
* `Occupation`              
* `Cat_ciudad`             : Categoría de la ciudad de residencia del usuario
* `Years_en _la_ciudad`    : Número de años que el usario lleva residiendo en la ciudad
* `Estado_civil`           : Casado(1), soltero(0)
* `Product_Category_1`       : Categoría 1 del producto 
* `Product_Category_2`     : Categoría 2 del producto
* `Product_Category_3`     : Categoría 3 del producto
* `Imp_Compra`             : Importe o precio del producto

Por otro lado, el dataset `BlackFriday_Products.xlsx` contiene la descripción de cada uno de los productos junto con su valoración en un rango del 1 al 5.

* `Product`                : Código de producto
* `Descripción`            : Descripción del producto
* `average_product_rating` : valoración media del producto en un rango del 1 al 5


Carga los datos de los datasets  `BlackFriday.csv` y `BlackFriday_Products.xlsx` en dos dataframes llamados `ventas` y `productos` respectivamente.

### 2.  Calcular los primeros estadísticos `mínimo`, `máximo`, `media` y `desviación típica` del importe de compra  `Imp_Compra`.


```python
# Con la función describe calculamos todos los estadisticos:

round(ventas.Imp_Compra.describe(),2)
```

```python
# También lo podriamos haber calculado de forma individual:

print("La media es :",round(ventas.Imp_Compra.mean(),2))
print("El máximo es :",round(ventas.Imp_Compra.max(),2))
print("El mínimo es :",round(ventas.Imp_Compra.min(),2))
print("La desviacion tipíca es :",round(ventas.Imp_Compra.std(),2))
```

###  5 Selecciona a los usuarios que han realizado alguna compra con importe por debajo de la media. Muestra los resultados sin repetición de usuarios.


```python
# Vemos cual es la media de las compras (9333.86)

media = ventas.Imp_Compra.mean()

media

# Seleccionamos los importes inferiores a la media
# Observamos que todos los importes son inferiores a 9333.86

compra_media = ventas[ventas.Imp_Compra < media]

compra_media

# No queremos ningún usuario repetido

compra_media_2 = compra_media[compra_media.User_ID.duplicated()]

compra_media_2


```

### 11 Construye un dataframe llamado `resumen` que muestre la información de código de usuario, código del producto que ha comprado, la valoración del producto  y el importe gastado.


```python
# Creamos el  dataframe con un merge con un inner. 


resumen1 = pd.merge(ventas, productos,  left_on = "Product_ID", right_on = "Product", how = "inner")

resumen1
```

```python
# Seleccionamos las columnas:

resumen = resumen1[["User_ID", "Product_ID", "average_product_rating", "Imp_Compra"]]

resumen
```
